##### Dictionary Assignment

```python
# Question 1: Create a dictionary of student names and their ages

student_dict = {'Arun':21, 
                'Kiran':22, 
                'Pavan':22,
                'Prasad':21,
                'Manas':22}
print(student_dict)
```

```python
# Question 2: Add a new key-value pair to dictionary {'a': 1, 'b': 2, 'c': 3}

dictionary = {'a': 1, 'b': 2, 'c': 3}
dictionary['d']= 4 #1 
dictionary.update({'e': 5}) #2 update()
print(dictionary)
```

```python
# Question 3: Get all keys from dictionary {'name': 'John', 'age': 25, 'city': 'New York'}

dictionary = {'name': 'John', 'age': 25, 'city': 'New York'}
print(dictionary.keys())
```

```python
# Question 4: Get all values from dictionary {'python': 3, 'java': 2, 'c++': 1}

dictionary = {'python': 3, 'java': 2, 'c++': 1}
print(dictionary.values())
```

```python
# Question 5: Check if key 'age' exists in {'name': 'Alice', 'age': 30, 'city': 'London'}

dictionary = {'name': 'Alice', 'age': 30, 'city': 'London'}

if 'age' in dict:
    print('exists')
else:
    print('not exists')
```

```python
# Question 6: Remove key 'temp' from {'a': 1, 'b': 2, 'temp': 3, 'c': 4}

dictionary = {'a': 1, 'b': 2, 'temp': 3, 'c': 4}
remove_key = dictionary.pop('temp')
print(remove_key)
print(dictionary)
```

```python
# Question 7: Find the sum of all values in {'math': 85, 'science': 92, 'english': 78}

dictionary = {'math': 85, 'science': 92, 'english': 78}
print(sum(dictionary.values()))
```

```python
# Question 8: Create a dictionary with squares of numbers 1 to 5

dictionary = {}

for i in range(1,6):
    dictionary[i] = i*i
print(dictionary)
```

```python
# Question 9: Count frequency of each character in string 'hello'

string = 'hello'
frequency = {}
for ch in string:
    if ch in frequency:
        frequency[ch] += 1
    else:
        frequency[ch] = 1
print(frequency)
```

```python
# Question 10: Merge two dictionaries {'a': 1, 'b': 2} and {'c': 3, 'd': 4}

dictionary1 = {'a': 1, 'b': 2}
dictionary2 = {'c': 3, 'd': 4}
dictionary1.update(dictionary2)
print(dictionary1)
```

```python
# Question 11: Create a nested dictionary: {'person': {'name': 'Alice', 'age': 25}}

nested_dict = {}
nested_dict['person'] = {'name': 'Alice', 'age': 25}

print(nested_dict)
```

```python
# Question 12: Access nested value 'name' from {'person': {'name': 'Alice', 'age': 25}}

print(nested_dict['person']['name'])
```

```python
# Question 13: Create a dictionary with list values: {'fruits': ['apple', 'banana'], 'colors': ['red', 'blue']}

dict_with_list = {}
dict_with_list['fruits'] = ['apple', 'banana']
dict_with_list['colors'] = ['red', 'blue']
print(dict_with_list)
```

```python
# Question 14: Add 'orange' to the 'fruits' list in nested dictionary

dict_with_list['fruits'].append('orange')
print(dict_with_list)
```

```python
# Question 15: Create a dictionary with tuple values: {'coordinates': (10, 20), 'rgb': (255, 0, 0)}

dict_with_tuple = {}
dict_with_tuple['coordinates'] = 10,20
dict_with_tuple['rgd'] = 255, 0, 0
print(dict_with_tuple)
```

```python
# Question 16: Extract first coordinate from nested tuple

print(dict_with_tuple.get('coordinates'))
```

```python
# Question 17: Create a dictionary with set values: {'vowels': {'a', 'e', 'i'}, 'consonants': {'b', 'c', 'd'}}

dict_with_set = {}
dict_with_set['vowels'] = {'a', 'e', 'i'}
dict_with_set['consonents'] = {'b', 'c', 'd'}
print(dict_with_set)
```

```python
# Question 18: Add 'o' to vowels set in nested dictionary

dict_with_set['vowels'].add('o')
print(dict_with_set)
```

```python
# Question 19: Create a 3-level nested dictionary: {'company': {'department': {'employee': {'name': 'John', 'id': 123}}}}

third_level_dict = {}

third_level_dict['company'] = {}
third_level_dict['company']['department'] = {}
third_level_dict['company']['department']['employee'] = {}
third_level_dict['company']['department']['employee']['name'] = 'John'
third_level_dict['company']['department']['employee']['id'] = 123

print(third_level_dict)
```

```python
# Question 20: Access employee name from 3-level nested dictionary

print(third_level_dict['company']['department']['employee']['name'])
```
